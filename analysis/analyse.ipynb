# Data Set analysis 

## Correlation assesement
Let's look at the plot distibution of our data features

```python
features = list(train_data.columns)
```

```python
season_sums = set()
season_avg = set()
season_std = set()

for f in features:
    if '_season_sum' in f:
        season_sums.add(f)
    elif '_season_std' in f:
        season_std.add(f)
    elif '_season_average' in f:
        season_avg.add(f)
```

```python
print(len(season_avg), len(season_std), len(season_sums))
```

```python
common_vars = set()

for f in season_sums:
    
    temp =  f[:len(f)-11]
    
    for e in season_std:
        if temp in e:
            common_vars.add(temp)
            break
        
        
len(common_vars)
```
